## 1) Imports necessários

```python
from pyspark.sql.functions import *
from delta.tables import *
from datetime import datetime
```

## 2) Definindo catálogo e o banco de dados que a tabela deve ser salva

```python
catalog_name = 'projetos_jr'
database_name = 'camada_bronze'
table_name = 'br_teste'
```

##3) Salvando o caminho da pasta

```python
path_csv = '/Workspace/Users/thaiscarvalheiro17@gmail.com/dados_com_problemas_qualidade.csv'
```

## 4) Criando tabela spark

```python
df = spark.read.format('csv').option('header', 'true').load(path_csv)
```

```python
df = df.withColumn(
    "dt_processamento",
    current_timestamp()
)

df = df.withColumn(
    "arquivo_origem",
    col("_metadata.file_path")
)
```

##5) Salvando dentro do ambiente analítico

```python
try:
    df.write.mode("append").saveAsTable(f'{catalog_name}.{database_name}.{table_name}')
except Exception as e:
    print(f"Erro no processamento: {e}")
                                                                                                                                        
```
